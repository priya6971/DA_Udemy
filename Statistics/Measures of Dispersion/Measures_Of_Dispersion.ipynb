Import all the required libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Load the data

```python
data = pd.read_csv('/content/process_engineering_data.csv')
```

```python
data.head()
```

**Measures Of Dispersion**

1. Range

```python
temp_range = data['temperature'].max() - data['temperature'].min()

print(temp_range)
```

2. IQR

```python
Q1 = data['temperature'].quantile(0.25)
Q3 = data['temperature'].quantile(0.75)

temp_iqr = Q3 - Q1
print(temp_iqr)
```

3. Variance and Standard Deviation

```python
pressure_variance = data['pressure'].var()
pressure_std = data['pressure'].std()

print("The variance of a pressure column is: ", pressure_variance)
print("The standard deviation of a pressure column is: ", pressure_std)
```

4. Skewness

```python
flow_rate_skewness = data['flow_rate'].skew()
print(flow_rate_skewness)
```

5. Kurtosis

```python
from scipy.stats import kurtosis

np.random.seed(42)
portfolio_returns = np.random.normal(loc=0, scale=1, size=1000)  # Normal distribution
portfolio_returns = np.append(portfolio_returns, np.random.normal(loc=0, scale=5, size=50))  # Adding extreme values

# Calculate kurtosis
kurt_value = kurtosis(portfolio_returns)
print(f"Kurtosis: {kurt_value:.2f}")

# Plotting the distribution
sns.histplot(portfolio_returns, bins=30, kde=True)
plt.title(f'Portfolio Returns Distribution (Kurtosis: {kurt_value:.2f})')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.show()
```
