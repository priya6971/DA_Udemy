Import all the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Generate and plot a weight data

```python
# Generate weights data (in kg)
weights = np.random.normal(loc=70, scale=15, size=1000)  # Mean=70kg, Std=15kg

# Plot histogram
plt.hist(weights, bins=15, edgecolor='black')
plt.title('Histogram of Weights')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

```python
from scipy.stats import norm

# Fit a normal distribution to the data
mu, std = norm.fit(weights)

# Plot histogram and PDF
plt.hist(weights, bins=15, density=True, alpha=0.6, color='b', edgecolor='black')
x = np.linspace(40, 100, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = f"Fit results: mu = {mu:.2f},  std = {std:.2f}"
plt.title(title)
plt.xlabel('Weight (kg)')
plt.ylabel('Density')
plt.show()
```

```python
# Parameters for standard normal distribution
mu_standard = 0
sigma_standard = 1

# Generate data
x_standard = np.linspace(mu_standard - 3*sigma_standard, mu_standard + 3*sigma_standard, 100)
pdf_standard = norm.pdf(x_standard, mu_standard, sigma_standard)

# Plot
plt.plot(x_standard, pdf_standard, label='Standard Normal Distribution')
plt.title('Standard Normal Distribution ($\mu=0$, $\sigma=1$)')
plt.xlabel('x')
plt.ylabel('Density')
plt.legend()
plt.show()
```
