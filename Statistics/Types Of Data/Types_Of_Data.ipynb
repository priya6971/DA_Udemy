Import all the required libraries

```python
import numpy as np
import pandas as pd
```

Create the nominal and ordinal type of data

```python
np.random.seed(0)
genders = ['male', 'female'] # nominal data
education_level = ['PhD', 'M.Tech', 'B.Tech', '12th'] # ordinal data

num_samples = 1000
gender_data = np.random.choice(genders, size=num_samples)
education_data = np.random.choice(education_level, size=num_samples)

types_of_data_df = pd.DataFrame({
    'Nominal Data': gender_data,
    'Ordinal Data': education_data
})

types_of_data_df
```

Create the interval and ratio data

```python
temp_data = np.random.randint(0, 100, size=num_samples) # tenperature
income_data = np.random.randn(num_samples) # income

# Ensure that all the income data values should be ideally positive (1000$)
income_data_positive = income_data - income_data.min() + 1

types_of_data_df['Interval Data'] = temp_data
types_of_data_df['Ratio Data'] = income_data_positive

types_of_data_df
```

Create the discrete and the continous data

```python
num_of_children_data = np.random.randint(0, 10, size=num_samples)
weight_data = np.random.rand(num_samples) * 10

types_of_data_df['Discrete Data'] = num_of_children_data
types_of_data_df['Continous Data'] = weight_data

types_of_data_df
```
