**Import all the required libraries**

```python
import pandas as pd
import numpy as np
```

**Load the data**

```python
dataframe = pd.read_csv('/content/drive/MyDrive/UDEMY Course Recordings/Statistics Module/Sampling Techniques /Population_Survey_Data.csv')
dataframe.head()
```

```python
dataframe.shape
```

**1. Random Sampling**

```python
# sample of size = 100 from the 1000 records in the population data
random_sample = dataframe.sample(n = 100, random_state=1)
print(random_sample.head())
```

```python
random_sample.shape
```

**2. Stratified Sampling**

```python
stratified_sampling = dataframe.groupby('Region', group_keys=False).apply(lambda x : x.sample(min(len(x), 25)))

stratified_sampling.to_csv('stratified_sampling.csv')
```

```python
stratified_sampling
```

```python
dataframe_manufacturing = pd.read_csv('/content/drive/MyDrive/UDEMY Course Recordings/Statistics Module/Sampling Techniques /Manufacturing_Data.csv')
dataframe_manufacturing.head()
```

**3. Cluster Sampling**

```python
clusters = dataframe_manufacturing['BatchNumber'].unique()
clusters
```

```python
selected_clusters = np.random.choice(clusters, size=5, replace=False)
selected_clusters
```

```python
clustered_samples = dataframe_manufacturing[dataframe_manufacturing['BatchNumber'].isin(selected_clusters)]
clustered_samples
```

```python
clustered_samples.to_csv('clustered_samples.csv')
```

**4. Systematic Sampling**

```python
systematic_sampling = dataframe.iloc[::10, :]
systematic_sampling
```

**For the entire population, let's try to compute the measures of central tendency: Mean, Median and Mode**

```python
mean_income = dataframe['Income'].mean()
median_income = dataframe['Income'].median()

print(f"Mean Income: {mean_income}")
print(f"Median Income: {median_income}")
```

```python
mode_income = dataframe['Income'].mode()[0]

print(f"Mode Income: {mode_income}")
```
